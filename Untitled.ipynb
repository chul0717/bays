```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt

%matplotlib inline
plt.style.use('ggplot')
```

```python
siteA = np.loadtxt('data/siteA.txt')
siteB = np.loadtxt('data/siteB.txt')
```

```python
x = np.linspace(0, 1, 1000)
x
```

```python
y = stats.beta(1, 1).pdf(x)
```

```python
def plot_with_fill(ax, x, y, label, fill=True):
    lines = ax.plot(x, y, label=label, lw=2)
    if fill:
        ax.fill_between(x, 0, y, alpha=0.2, color=lines[0].get_c())
    ax.legend(loc="upper right")
```

```python
fig, ax = plt.subplots()
plot_with_fill(ax, x, y, label = 'uniform_CTR')
```

```python
first_50 = siteA[:50]
first_50
```

```python
def make_beta(data, idx):
    l = []
    for i in idx:
        data1 = data[:i]
        alpha = list(data1).count(1)
        beta = list(data1).count(0)
        print(alpha, beta, i)
        l.append(stats.beta(a=1+alpha, b=1+beta))
    d = {key: value for key, value in zip(idx, l)}
    return d
```

```python
make_beta(siteA, idx)
```

```python
y1 = make_beta(first_50).pdf(x)
```

```python
fig, ax = plt.subplots(figsize=(16, 8))
plot_with_fill(ax, x, y, label = 'Prior')
plot_with_fill(ax, x, y1, label = 'Posterior after 50 views')

```
